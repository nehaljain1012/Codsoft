Preprocessing Data

```python
df_encoder = LabelEncoder()
new_col = df_encoder.fit_transform(ds_combined["merchant"].values)
ds_combined["merchant_new"] = new_col
ds_combined.drop(labels = ["merchant"], axis = 1, inplace = True)
```

```python
df_encoder = LabelEncoder()
new_coll = df_encoder.fit_transform(ds_combined["category"].values)
ds_combined["category_new"] = new_coll
ds_combined.drop(labels = ["category"], axis = 1, inplace = True)
```

```python
ds_combined = pd.get_dummies(ds_combined)
ds_combined.drop(labels=['gender_F'], axis = 1, inplace = True)
```

```python
ds_combined.head()
```

```python
ds_combined.tail()
```

```python
x = ds_combined.drop("is_fraud", axis = 1)
y = ds_combined["is_fraud"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

Training Logistic Regression Model

```python
logistic_model=LogisticRegression()
```

```python
logistic_model.fit(x_train, y_train)
```

```python
logistic_prediction = logistic_model.predict(x_test)
```

print("Logistic Regression Model: ")
print(confusion_matrix(y_test, logistic_prediction))
print(classification_report(y_test, logistic_prediction))
print("Accuracy: ", accuracy_score(y_test, logistic_prediction))

Training Decision Tree Model

```python
decision_tree_model=DecisionTreeClassifier()
```

```python
decision_tree_model.fit(x_train, y_train)
```

```python
decision_tree_predictions = decision_tree_model.predict(x_test)
```

```python
print("Decision Tree Model: ")
print(confusion_matrix(y_test, decision_tree_predictions))
print(classification_report(y_test, decision_tree_predictions))
print("Accuracy: ", accuracy_score(y_test, decision_tree_predictions))
```

Training Random Forest Model

```python
random_forest_model = RandomForestClassifier(n_estimators=100, n_jobs=-1, random_state=42)
```

```python
random_forest_model.fit(x_train, y_train)
```

```python
random_forest_predictions = random_forest_model.predict(x_test)
```

```python
print("Random Forest Model: ")
print(confusion_matrix(y_test, random_forest_predictions))
print(classification_report(y_test, random_forest_predictions))
print("Accuracy: ", accuracy_score(y_test, random_forest_predictions))
```
